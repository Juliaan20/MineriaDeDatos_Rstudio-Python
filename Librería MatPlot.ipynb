```python
def f(x):
    return np.exp(-x)*np.cos(2**np.pi)
```

```python
x1 = np.arange(0,5.0,0.1)
x2=np.arange(0,5.0,0.2)
```

```python
x1
```

```python
x2
```

```python
plt.figure(1)
plt.subplot(211)
plt.plot(x1, f(x1), 'ro', x2, f(x2),'k')
plt.subplot(212)
plt.plot(x2, f(x2), 'g--')
plt.show()
```
